```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [18, 22, 25, 30, 35, 40, 45, 50]
kiloListesi = [60, 65, 70, 75, 50, 85, 60, 95]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi,color="red")
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi,"y")
```

```python
plt.plot(numpyKiloListesi, numpyYasListesi,"b")
plt.xlabel("Kilo")
plt.ylabel("Yaş")
plt.title("Yaş-Kilo Grafiği")
plt.show()
```

```python
numpydizisi1 = np.linspace(0, 10, 50)
```

```python
numpydizisi1
```

```python
numpydizisi2 = numpydizisi1 ** 2
```

```python
numpydizisi2
```

```python
plt.plot(numpydizisi1,numpydizisi2,"g--")
```

```python
plt.plot(numpydizisi1,numpydizisi2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpydizisi1,numpydizisi2,"r*-")
plt.xlabel("x")
plt.ylabel("y")
plt.title("1.grafik")
plt.subplot(1,2,2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2.grafik")
plt.plot(numpydizisi2,numpydizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.4,0.2,0.3,0.3])
figureAxes.plot(numpydizisi1,numpydizisi2,"r--")
figureAxes.set_xlabel("x")
figureAxes.set_ylabel("y")
figureAxes.set_title("Başlık")
```

```python
figure2 = plt.figure()
axes1 = figure2.add_axes([0.1,0.1,0.8,0.8])
axes2 = figure2.add_axes([0.2,0.49,0.3,0.3])
axes1.plot(numpydizisi1,numpydizisi2,"b--")
axes1.set_xlabel("x")
axes1.set_ylabel("y")
axes1.set_title("Büyük Grafik")
axes2.plot(numpydizisi2,numpydizisi1,"r-")
axes2.set_xlabel("x")
axes2.set_ylabel("y")
axes2.set_title("Küçük Grafik")
```

```python
(benimFigure4, benimEksen1) = plt.subplots(nrows=1, ncols=2)
type(benimEksen1)
for eksen in benimEksen1:
    eksen.plot(numpydizisi1,numpydizisi2,"g--")
    eksen.set_xlabel("x")
    eksen.set_ylabel("y")
plt.tight_layout() # grafiklerin birbirine girmesini engeller
```

```python
fig = plt.figure(figsize=(8,8), dpi=100)  # figsize ile figürün boyutunu, dpi ile çözünürlüğünü belirler
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.plot(numpydizisi1, numpydizisi2, label = "numpydizisi ** 2")
ax.plot(numpydizisi1, numpydizisi1, label = "numpydizisi ** 1")
ax.legend(loc=2)  # legend konumunu belirler
```

```python
fig.savefig("grafik.png", dpi=100)  # grafiği kaydeder
```
