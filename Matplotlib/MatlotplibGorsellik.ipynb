```python
import numpy as np
import matplotlib.pyplot as plt
# Jupyter Notebook'ta grafiklerin hücre içinde görüntülenmesini sağlar
%matplotlib inline
```

```python
numpydizisi1 = np.linspace(0, 10, 20)
```

```python
numpydizisi1
```

```python
numpydizisi12 = numpydizisi1 ** 2
```

```python
(benimFigur , benimEksen) = plt.subplots(figsize=(8,8), dpi=100)  # figsize ile figürün boyutunu, dpi ile çözünürlüğünü belirler
benimEksen.plot(numpydizisi1, numpydizisi12, "g",alpha=0.5) # alpha ile çizginin saydamlığını belirler
benimEksen.plot(numpydizisi1, numpydizisi1, "r",linewidth=2) # linewidth ile çizginin kalınlığını belirler
benimEksen.plot(numpydizisi12, numpydizisi12, "b",linestyle="--") # alpha ile çizginin saydamlığını belirler
benimEksen.plot(numpydizisi12, numpydizisi12*2, "black",linestyle="--",marker="o",markerfacecolor="red") # marker ile nokta şekillerini belirler, markersize ile nokta boyutunu belirler
benimEksen.plot(numpydizisi12, numpydizisi12*3, "yellow",linestyle="--",marker="o",markerfacecolor="red",markersize=10) # marker ile nokta şekillerini belirler, markersize ile nokta boyutunu belirler
```

scatter

```python
plt.scatter(numpydizisi1, numpydizisi12) # scatter ile nokta grafiği oluşturur
```
