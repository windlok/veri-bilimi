Serial/Seriler

```python
benimSozlugum = {"bir": 1, "iki": 2, "uc": 3}
benimSerim = pd.Series(benimSozlugum)
benimSerim
```

```python
benimListem = [1, 2, 3, 4, 5]
benimisimlerim = ["bir", "iki", "uc", "dort", "bes"]
benimSerim2 = pd.Series(data=benimListem, index=benimisimlerim)
benimSerim2
```

```python
numpyDizim = np.array([1, 2, 3, 4, 5])
benimSerim3 = pd.Series(numpyDizim)
benimSerim3
```
