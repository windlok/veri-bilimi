```python
yenisozlugu = {
    "departman" : ["yazılım","satış","pazarlama","yazılım","hukuk","pazarlama","yazılım","yazılım"],
    "calisanlar": ["ali","veli","ayşe","fatma","mehmet", "ayşe", "fatma", "mehmet"],
    "maas": [50000, 40000, 30000, 45000, 35000, 30000, 45000, 35000]
}
```

```python
yenicalisanlar = pd.DataFrame(yenisozlugu)
yenicalisanlar
```

```python
grupobjesi = yenicalisanlar.groupby("departman")
```

```python
grupobjesi.count()
```

```python
grupobjesi.mean(numeric_only=True)   # veya
```

```python
grupobjesi.min()
```
