```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

MinMaxScaler ile Veri Ölçeklendirme

**fit() : ölçeklemede belirli bir verinin ortalamasını ve std sapmasını hesaplamak için kulllanılır.
**transform() :fit() yönetimi ile hesaplanan ortalama ve std sapma kullanılarak ölçekleme gerçekleştirmek için kullanılır.
**the fit_transform() hem fit hemde transform eder 

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
```

```python
X_train_scaled = mm_scaler.fit_transform(X_train)
X_test_scaled
```

```python
mm_scaler.data_max_
```

```python
mm_scaler.scale_
```

```python
X_train_scaled[:5]
```

```python
np.set_printoptions(precision=3, suppress=True)
```

```python
X_train_scaled[:5]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score ,mean_absolute_error
model = LinearRegression()
model.fit(X_train_scaled, Y_train)
```

```python
print("katmayılar : ", model.coef_)
print("kesme terimi :"  , model.intercept_)
print("katsayi sayisi : ", model.coef_.shape)
```

```python
x_test_scaled = mm_scaler.transform(X_test)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
y_pred
```

```python
Y_test
```

Modelin Değerlendirilmesi

R^2: Determinasyon katsayısı. score() fonksiyonu determinasyon katsayısı verir.

```python
model.score(x_test_scaled, Y_test)
```

```python
r2_score(Y_test, y_pred)
```

```python
mse = mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
print("MSE : ", mse)
print("RMSE : ", rmse)
```
