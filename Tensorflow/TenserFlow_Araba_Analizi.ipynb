model egitme kısmı

```python
y = dataFrame['price'].values
x = dataFrame.drop('price',axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sclaer = MinMaxScaler()
```

```python
x_train = sclaer.fit_transform(x_train)
```

```python
x_test = sclaer.transform(x_test) 
```

```python
from tensorflow.keras.models import Sequential #yapay sinir ağı modeli oluşturmak için kullanılır
from tensorflow.keras.layers import Dense #yapay sinir ağı katmanları oluşturmak için kullanılır
```

```python
x_train.shape
```

```python
model =Sequential()

model.add(Dense(12,activation='relu')) #giriş katmanı
model.add(Dense(12,activation='relu')) #giriş katmanı
model.add(Dense(12,activation='relu')) #giriş katmanı
model.add(Dense(12,activation='relu')) #giriş katmanı

model.add(Dense(1)) #çıkış katmanı

model.compile(optimizer='adam',loss='mse') #optimizer: modelin öğrenme sürecini optimize etmek için kullanılan algoritma
```

modeli eğitmeye basladıgımız ana kısım

```python
#validation_data: modelin doğruluğunu değerlendirmek için kullanılan veri seti
#batch_size: modelin eğitimi sırasında kullanılacak veri miktarı
#epochs: modelin eğitimi sırasında kaç kez tüm veri seti üzerinde çalışılacağını
model.fit(x = x_train, y = y_train,validation_data=(x_test,y_test),batch_size=32,epochs=150) #modeli eğitmek için kullanılır
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
tahminler = model.predict(x_test)
```

```python
tahminler
```

```python
mean_absolute_error(y_test,tahminler)
```
