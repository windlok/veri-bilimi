```python
y = dataFrame["Type"].values
x = dataFrame.drop(["Type"], axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x = x_train, y = y_train, validation_data=(x_test,y_test), batch_size=128, epochs=600, callbacks=[EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model1 = Sequential()

model1.add(Dense(units=30,activation="relu"))
model1.add(Dense(units=15,activation="relu"))
model1.add(Dense(units=15,activation="relu"))
model1.add(Dense(units=1,activation="sigmoid"))

model1.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlystopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model1.fit(x_train,y_train,epochs=600,validation_data=(x_test,y_test),callbacks=[earlystopping])
```

```python
modelKaybi1= pd.DataFrame(model1.history.history)
```

```python
modelKaybi1.plot()
```

```python
model2 = Sequential()

model2.add(Dense(units=30,activation="relu"))
model2.add(Dropout(0.6)) #dropout şu anlama geliyor, modelin bazı nöronlarını rastgele kapatmak. Bu sayede modelin genelleme yeteneği artar ve overfitting (aşırı öğrenme) sorununu azaltır.

model2.add(Dense(units=15,activation="relu"))
model2.add(Dropout(0.6))

model2.add(Dense(units=15,activation="relu"))
model2.add(Dropout(0.6))

model2.add(Dense(units=1,activation="sigmoid"))
model2.add(Dropout(0.6))

model2.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model2.fit(x_train,y_train,epochs=600,validation_data=(x_test,y_test),callbacks=[earlystopping])
```

```python
modelKaybi2 = pd.DataFrame(model2.history.history)
```

```python
modelKaybi2.plot()
```

```python
probas = model2.predict(x_test)
tahmiblerimiz = (probas > 0.5).astype(int).reshape(-1,)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
classification_report(y_test,tahmiblerimiz)
```

```python
print(confusion_matrix(y_test,tahmiblerimiz))
```
