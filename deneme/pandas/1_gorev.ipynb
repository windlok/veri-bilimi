```python
import numpy as np
import pandas as pd
```

```python
isim = input("Lütfen isim gitiniz: ")
sayi = input("Lütfen sayı giriniz: ")

isi = isim.split() 
sa = sayi.split() 
say = list(map(int, sa))

benimListem = pd.Series(data=say, index=isi)
print(benimListem)
print(f'Boyut Sayisi: {np.ndim(benimListem)}')
print(f'Eleman Sayisi: {np.size(benimListem)}')
print(f'sekil: {np.shape(benimListem)}')

```

```python
# Sadece veri sütununa 'sirasi' adını verdik.
dataFram = pd.DataFrame(benimListem, columns=['sirasi']) 
print(dataFram)
```
