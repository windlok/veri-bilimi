```python
import pandas as pd
```

```python
veri = input("kaç kişi almak iştiyorsun")
##dict yapsııs ile yapılacaktır
isimler_listesi = []
notlar_listesi = []

isim = {
    'isimler': isimler_listesi,
    "notlar": notlar_listesi
}
veri = int(veri)
for i in range(veri):
    ism = input("isim girinnizi ve not giriniz boşluk birakarak")
    iss = ism.split()
    isimler_listesi.append(iss[0])
    notlar_listesi.append(int(iss[1]))
tablo = pd.DataFrame(isim)
print(tablo)
```
