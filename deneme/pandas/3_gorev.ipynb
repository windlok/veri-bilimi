```python
import pandas as pd
```

```python
id_listesi = []
isimler_listesi = []
maas_listesi = []
personel = {
    "idler": id_listesi,
    "isimler": isimler_listesi,
    "maaslar": maas_listesi
}

adet = input("kaç kullanici isteniyor")
adet_int = int(adet)

for n in range(adet_int):
    perso = input("id, isimleri, maasi giriniz:")
    person = perso.split()
    id_listesi.append(int(person[0]))
    isimler_listesi.append(person[1])
    maas_listesi.append(int(person[2]))

tablo = pd.DataFrame(personel)
tablo = tablo.set_index('idler')
print(tablo)

```
