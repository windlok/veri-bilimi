```python
import pandas as pd
idi_liste = []
isim_liste = []
departman_liste = []
maas_liste = []
personal = {
    "Id": idi_liste,
    "isim": isim_liste,
    "Departman": departman_liste,
    "Maas": maas_liste
}
input1 = input("Kaç adet personel gireceksiniz: ")
for i in range(int(input1)):
    input2 = input("id, isim, departman, maas bosluk bıkarak yaz")
    veri = input2.split()
    idi_liste.append(int(veri[0]))
    isim_liste.append(veri[1])
    departman_liste.append(veri[2])
    maas_liste.append(int(veri[3]))
veriDataFrame = pd.DataFrame(personal)
veriDataFrame = veriDataFrame.set_index("Id")

son_girdi = input("Aranan ID ve Maaş barajını boşluk bırakarak giriniz: ")
son_veriler = son_girdi.split()
aranan_id = int(son_veriler[0])
maas_baraji = int(son_veriler[1])

print("\n--- ILK 2 PERSONEL ---")
print(veriDataFrame.iloc[:2])

print("\n--- ARANAN KISI ---")
if aranan_id in veriDataFrame.index:
    print(veriDataFrame.loc[aranan_id])
else:
    print("Aranan ID bulunamadı.")

print(f"\n--- MAASI {maas_baraji} UZERI OLANLAR ---")

print(veriDataFrame[ veriDataFrame['Maas'] > maas_baraji ])
```
