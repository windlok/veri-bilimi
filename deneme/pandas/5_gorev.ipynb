```python
import pandas as pd 
```

```python
gun_listesi = []
sicaklik_listesi = []
takvim = {
    "gunler": gun_listesi,
    "sıcaklık": sicaklik_listesi
}
input1 = int(input("kaç adet hava tahmini eklenecek: "))

for i in range(input1):
    input2 = input("gün ve sıcaklığı giriniz bosluk bırakarak: ")
    input3 = input2.split()

    gun_listesi.append(input3[0])
    sicaklik_listesi.append(float(input3[1]))

veridataframe = pd.DataFrame(takvim)    

ortalama = veridataframe["sıcaklık"].mean()
veridataframe["sıcaklık"] = veridataframe["sıcaklık"].fillna(ortalama)
print(veridataframe)
```
