```python
import pandas as pd
```

```python
katagori_listesi = []
fiyat_listesi = []

veri = {
    "kategori": katagori_listesi,
    "fiyat": fiyat_listesi
}

satis_adeti = int(input("Kaç adet ürün satıldı? "))

for i in range(satis_adeti):
    kata = input(f"{i+1}. ürünün katagorisini giriniz: ")
    kate = kata.split()
    katagori_listesi.append(kate[0])
    fiyat_listesi.append(float(kate[1]))

veriler = pd.DataFrame(veri)

print(veriler.groupby("kategori").sum())
```
