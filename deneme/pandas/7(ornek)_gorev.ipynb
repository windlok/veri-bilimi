```python
import pandas as pd

# 1. PERSONEL SAYISINI AL
n = int(input("Personel sayısını giriniz: "))

# 2. İK TABLOSUNU OLUŞTUR (ID ve İsim)
id_listesi_ik = []
isim_listesi = []

print(f"\n--- Şimdi {n} adet ID ve İsim giriniz ---")
for i in range(n):
    veri = input().split()
    id_listesi_ik.append(int(veri[0]))
    isim_listesi.append(veri[1])

df_ik = pd.DataFrame({"ID": id_listesi_ik, "Isim": isim_listesi})

# 3. MUHASEBE TABLOSUNU OLUŞTUR (ID ve Maaş)
id_listesi_muh = []
maas_listesi = []

print(f"\n--- Şimdi {n} adet ID ve Maaş giriniz ---")
for i in range(n):
    veri = input().split()
    id_listesi_muh.append(int(veri[0]))
    maas_listesi.append(int(veri[1]))

df_muh = pd.DataFrame({"ID": id_listesi_muh, "Maas": maas_listesi})

# 4. BÜYÜK BİRLEŞME (Merge)
# 'on' parametresi iki tablodaki ortak sütunu belirtir
sonuc = pd.merge(df_ik, df_muh, on="ID")

print("\n--- İK ve Muhasebe Birleşmiş Tablo ---")
print(sonuc)
```
