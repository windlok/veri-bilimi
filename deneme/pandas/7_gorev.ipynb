```python
import pandas as pd
```

```python
input1 = input("kaç adet sayı gireceksiniz: ")
id_liste = []
id_liste1 = []
maas_liste = []
isim_liste = []

tablo_ik = {"id": id_liste, "maas": maas_liste}
tablo_muhasebe = {"id": id_liste1,"isim": isim_liste}


for i in range((int(input1))*2):
    input2 = input("ister id ile maas , ister isim ile id gir ama bosluk birak ")
    input3 = input2.split()
    
    if input3[1].isdigit():
        id_liste.append(int(input3[0]))
        maas_liste.append(int(input3[1]))
    else:
        id_liste1.append(int(input3[0]))
        isim_liste.append(input3[1])

veridataframe = pd.DataFrame(tablo_ik)
veridataframe1 = pd.DataFrame(tablo_muhasebe)

birlesmis_tablo = pd.merge(veridataframe,veridataframe1,on="id")
print(birlesmis_tablo)
```
